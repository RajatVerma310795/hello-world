```
class Lecturer(object): 
    def __init__(self): 
        self.str1 = "lecturer"
        print ("Base1")
  
class SecGuard(object): 
    def __init__(self): 
        self.str2 = "SecGuard"        
        print ("Base2")
  
class Employee(Lecturer, SecGuard): 
    def __init__(self): 
          
        # Calling constructors of Base1 
        # and Base2 classes 
        Lecturer.__init__(self) 
        SecGuard.__init__(self) 
        print ("Derived")
          
    def printStrs(self): 
        print(self.str1, self.str2) 
ob = Employee() 
ob.printStrs()
```

```
def fooBar(): #decorator is a function that returns a function
  print("foobar")
  
```

```
def my_decorator(passedFunc):
  def wrapper_my_dec():
    print("Before the passed function")
    passedFunc()
    print("After the passed function")
    return wrapper_my_dec
    
```

```
@my_decorator
def greet():
  print("hi")
```

```
fooBar=my_decorator(fooBar)
```

```
def dataCleaner(func):
  def dataCleaner_wrapper(data, *args, **kwargs):
    distintList = list(set(data))
    filteredList = filter(lambda x: x not in [None, False,0], distintList)
    return func(filteredList, *args, **kwargs)
  return dataCleaner_wrapper
@dataCleaner
def myDataAnalysisFunction(data):
  return list(map(lambda datum: datum *5 , data))

retVal = myDataAnalysisFunction([1,6,"asd","asd",None, False])
print("Reurn value from decorated function is :", retVal)
```
